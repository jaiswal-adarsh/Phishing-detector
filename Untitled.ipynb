```python
tokenizer=RegexpTokenizer(r'[A-Za-z]+')
```

```python
phish_data.URL[0]
```

```python
#this will be pull letter which matches to expression
tokenizer.tokenize(phish_data.URL[0]) #using first row
```

```python
print('Getting words tokenized ...')
t0=time.perf_counter()
phish_data['text_tokenized']=phish_data.URL.map(lambda t:tokenizer.tokenize(t)) #performing all the rows
t1=time.perf_counter() -t0
print('Time taken',t1,'sec')
```

```python
phish_data.sample(5)
```

```python
#using snowball for stemming

stemmer=SnowballStemmer("english") #choosing a language
```

```python
print('Getting words stemmed ...')
t0=time.perf_counter()
phish_data['text_stemmed']=phish_data['text_tokenized'].map(lambda l:[stemmer.stem(word) for word in l])
t1=time.perf_counter()-t0
print('Time taken',t1,'sec')
```

```python
phish_data.sample(5)
```

```python
#joining all the listed words into sentence

print('Getting Joinwords ...')
t0=time.perf_counter()
phish_data['text_sent']=phish_data['text_stemmed'].map(lambda l:''.join(l))
t1=time.perf_counter()-t0
print('Time taken',t1,'sec')
```

```python
#creating a function to visualize the important keys from url using word cloud

def plot_wordcloud(text,mask=None,max_words=400,max_font_size=120,figure_size=(24.0,16.0),title=None,title_size=40,image_color=False):
    
    stopwords=set(STOPWORDS)
    more_stopwords={'con','http'}
    stopwords=stopwords.union(more_stopwords)
    
    wordcloud =WordCloud(background_color='white',
                        stopwords=stopwords,
                        max_words=max_words,
                        max_font_size=max_font_size,
                        random_state=42,
                        mask=mask)
    wordcloud.generate(text)
    
    plt.figure(figsize=figure_size)
    if image_color:
        image_colors=ImageColorGenerator(mask);
        plt.imshow(wordcloud.recolor(color_func=image_colors),interpolation="bilinear");
        plt.title(title,fontdict={'size':title_size,'verticalalignment': 'bottom'})
        
    else:
        plt.imshow(wordcloud);
        plt.title(title,fontdict={'size':title_size,'color':'green','verticalalignment':'bottom'})
        
    plt.axis('off');
    plt.tight_layout()
```

```python
data=good_sites.text_sent
data.reset_index(drop=True, inplace=True)
```

```python
common_text=str(data)
common_mask=np.array(Image.open('star.png'))

plot_wordcloud(common_text,common_mask,max_words=400,max_font_size=120,title='Most common words used in good URLS ',title_size=15)
```

```python
#visualizing internal links

#scraping website
#using cchrome webDriver
```

```python
browser=webdriver.Chrome(r"chromedriver.exe")
```

```python
list_urls=['https://www.flipkart.com/','https://www.flipkart.com/about-us'] #these are the phishing sites
links_with_text=[]
```

```python
#Using BeautifulSoup
#used for getting data out of HTML,XML and other markup language

for url in list_urls:
    browser.get(url)
    soup=BeautifulSoup(browser.page_source,"html.parser")
    
    for line in soup.find_all('a'):
        href=line.get('href')
        links_with_text.append([url,href])
```

```python
#turning url into data frames
#after getting the list of the website with hyperlinks turn them into a pandas DataFrame with columns "From (source)" to "to(destination)"
#scraping the hidden links
df=pd.DataFrame(links_with_text,columns=["from","to"])
```

```python
df
```

```python
#Drawing graph
#visualizing an internal link structure by feeding it to the network method from_pandas_edgelist first 
#and draw it by calling nx draw

GA=nx.from_pandas_edgelist(df,source="from",target="to")
nx.draw(GA,with_labels=False)
```

```python
#creating an object

cv=CountVectorizer()
```

```python
##transform all texts thyat are tokenized and stemmed
feature=cv.fit_transform(phish_data.text_sent)
```

```python
#converting sparse matrix into array to print transformed features
feature[:5].toarray()
```

```python
#Splitting the data
```

```python
trainX,testX,trainY,testY=train_test_split(feature,phish_data.Label)
```

```python
#using logistic regression (classification algorithm)
```

```python
#creating LR object
lr=LogisticRegression()
```

```python
lr.fit(trainX,trainY)
```

```python
lr.score(trainX,trainY)
```

```python
#LR is giving 76% accuracy, Now we will store the score in the dictionary to see which model is performing best
```

```python
Score_ml={}
Score_ml['Logistic Regression']=np.round(lr.score(testX,testY),2)
```

```python
print('Training Accuracy :',lr.score(trainX,trainY))
print('Training Accuracy :',lr.score(testX,testY))

con_mat=pd.DataFrame(confusion_matrix(lr.predict(testX),testY),
                     columns=['Predict :Bad','Predict :Good'],
                     index=['Actual :Bad','Actual :Good'])

print('\n Classification Report\n')
print(classification_report(lr.predict(testX),testY,
                           target_names=['Bad','Good']))

print('\n Confusion Matrix')
plt.figure(figsize=(6,4))
sns.heatmap(con_mat,annot=True,fmt='d',cmap="YlGnBu")

```

```python
 #using MultinomialNB (naive Bayes) to NLP problems. NB classifier algo is a family of probabilistic algo
```

```python
#creating MNB object
mnb=MultinomialNB()
```

```python
mnb.fit(trainX,trainY)
```

```python
mnb.score(testX,testY)
```

```python
#it is giving us 85% accuracy
```

```python
Score_ml['MultinomialNB']=np.round(mnb.score(testX,testY),2)
```

```python
print('Training Accuracy :',mnb.score(trainX,trainY))
print('Training Accuracy :',mnb.score(testX,testY))

con_mat=pd.DataFrame(confusion_matrix(mnb.predict(testX),testY),
                     columns=['Predict :Bad','Predict :Good'],
                     index=['Actual :Bad','Actual :Good'])

print('\n Classification Report\n')
print(classification_report(mnb.predict(testX),testY,
                           target_names=['Bad','Good']))

print('\n Confusion Matrix')
plt.figure(figsize=(6,4))
sns.heatmap(con_mat,annot=True,fmt='d',cmap="YlGnBu")
```

```python
acc=pd.DataFrame.from_dict(Score_ml,orient='index',columns=['Accuracy'])
sns.set_style('darkgrid')
sns.barplot(acc.index,acc.Accuracy)
```

```python
#So,Multinomiua; NB is the best fit algorithm, we will make Sklearn pipeline using multinomial NB
```

```python
pipeline_ls=make_pipeline(CountVectorizer(tokenizer=RegexpTokenizer(r'[A-Za-z]+').tokenize,stop_words='english'),MultinomialNB())
```

```python
trainX,testX,trainY,testY=train_test_split(phish_data.URL,phish_data.Label)
```

```python
pipeline_ls.fit(trainX,trainY)
```

```python
pipeline_ls.score(testX,testY)
```

```python
print('Training Accuracy :',pipeline_ls.score(trainX,trainY))
print('Training Accuracy :',pipeline_ls.score(testX,testY))

con_mat=pd.DataFrame(confusion_matrix(pipeline_ls.predict(testX),testY),
                     columns=['Predict :Bad','Predict :Good'],
                     index=['Actual :Bad','Actual :Good'])

print('\n Classification Report\n')
print(classification_report(pipeline_ls.predict(testX),testY,
                           target_names=['Bad','Good']))

print('\n Confusion Matrix')
plt.figure(figsize=(6,4))
sns.heatmap(con_mat,annot=True,fmt='d',cmap="YlGnBu")
```

```python
#saving this model using pickel

pickle.dump(pipeline_ls,open('phishing.pkl','wb'))
```
